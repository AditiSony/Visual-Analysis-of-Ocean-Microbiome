```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
from jupyter_dash import JupyterDash 
import dash 
from dash import dcc 
from dash import html
from dash.dependencies import Input, Output
```

## Taxonomy-wise Abundance

```python
# Reading taxonomy table
taxonomy= pd.read_csv('C:/Users/Aditi Sony/Downloads/Complex Data Visualisation/Summative Assignment/Data/Tara_OTUtableTax.csv')
taxonomy.head()
```

### Highest abundance Domain?

```python
Domain_group = taxonomy.groupby('Domain').sum()
Domain_group['Percent']= Domain_group.sum(axis=1, numeric_only=True) / 139
Domain_group.reset_index(inplace=True)
Domain_group = Domain_group.rename(columns = {'index':'Domain'})
```

```python
Domain_group.head()
```

```python
fig = px.pie(Domain_group, names="Domain", values= 'Percent', color= 'Domain')
fig.show()
```

### Highest abundance Phylum?

```python
Domain_Phylum= taxonomy.groupby(['Domain', 'Phylum']).agg(['sum'])
Domain_Phylum['Percent']= Domain_Phylum.sum(axis=1, numeric_only=True) / 139
Domain_Phylum.sort_values(by=['Percent'], inplace=True, ascending=False)
Domain_Phylum.reset_index(inplace=True)
Domain_Phylum.head()
```

```python
fig1 = px.bar(Domain_Phylum, x="Domain", y= 'Percent', color= 'Phylum', text_auto=True)
fig1.show()
```

### Abundance break-down by taxonomy hierarchy upto Genus level

```python
Taxonomy_groups= taxonomy.groupby(['Domain', 'Phylum', 'Class', 'Order', 'Family', 'Genus']).agg(['sum'])
Taxonomy_groups['Percent']= Taxonomy_groups.sum(axis=1, numeric_only=True) / 139
Taxonomy_groups.sort_values(by=['Percent'], inplace=True, ascending=False)
Taxonomy_groups.reset_index(inplace=True)
Taxonomy_groups.head()
```

```python
fig2 = px.sunburst(Taxonomy_groups, path=['Domain', 'Phylum', 'Class', 'Order', 'Family', 'Genus'], values='Percent', color= 'Genus')
fig2.show()
```

## Region-wise Abundance

```python
# Transposing The Genus Pivot Table
Genus= pd.read_csv('C:/Users/Aditi Sony/Downloads/Complex Data Visualisation/Summative Assignment/Data/PivotedTaxonomy/Tara_GenusTable.csv')
Genus= Genus.transpose()
Genus.columns = Genus.iloc[0]
Genus.drop(Genus.index[[0]], inplace=True)
Genus.reset_index(inplace=True)
Genus = Genus.rename(columns = {'index':'SampleID'})
Genus.head()
```

```python
# Reading Metadata file
Metadata= pd.read_csv('C:/Users/Aditi Sony/Downloads/Complex Data Visualisation/Summative Assignment/Data/Tara_SampleMeta.csv')
Metadata.head()
```

```python
# Joining Genus and Metadata files
Genus_Meta= pd.merge(Genus, Metadata, how= "outer", on=['SampleID', 'SampleID'])
Genus_Meta.shape
```

```python
Genus_Meta.head()
```

```python
Genus_Meta.nunique()
```

### Parallel Coordinate Sets Plot

```python
fig3 = px.parallel_categories(Genus_Meta, dimensions=['Layer of Origin', 'MarinePelagicBiomes', 'OceanAndSeaRegions', 'MarinePelagicProvince'])
fig3.show()
```

### Depth-wise Abundance

```python
Depth_group= Genus_Meta.drop(['Year', 'Month', 'SampleID'], axis=1)
Depth_group= Depth_group.groupby(['Layer of Origin']).sum()
Depth_group= Depth_group.transpose()
Depth_group.reset_index(inplace=True)
Depth_group = Depth_group.rename(columns = {'index':'Genus'})
Depth_group.head()
```

```python
Depth_melt = Depth_group.melt(id_vars='Genus')
fig4 = px.bar(Depth_melt, x="value", y="Layer of Origin", color='Genus', orientation='h')
fig4.show()
```

### Marine Pelagic Biome-wise Abundance

```python
Biome_group= Genus_Meta.drop(['Year', 'Month', 'SampleID'], axis=1)
Biome_group= Biome_group.groupby(['MarinePelagicBiomes']).sum()
Biome_group= Biome_group.transpose()
Biome_group.reset_index(inplace=True)
Biome_group = Biome_group.rename(columns = {'index':'Genus'})
Biome_group.head()
```

### Ocean-wise Abundance

```python
Ocean_group= Genus_Meta.drop(['Year', 'Month', 'SampleID'], axis=1)
Ocean_group= Ocean_group.groupby(['OceanAndSeaRegions']).sum()
Ocean_group= Ocean_group.transpose()
Ocean_group.reset_index(inplace=True)
Ocean_group = Ocean_group.rename(columns = {'index':'Genus'})
Ocean_group.head()
```

### Province-wise Abundance

```python
Province_group= Genus_Meta.drop(['Year', 'Month', 'SampleID'], axis=1)
Province_group= Province_group.groupby(['MarinePelagicProvince']).sum()
Province_group= Province_group.transpose()
Province_group.reset_index(inplace=True)
Province_group = Province_group.rename(columns = {'index':'Genus'})
Province_group.head()
```

### Sample-wise Abundance

```python
Sample_group= Genus_Meta.drop(['Year', 'Month', 'Layer of Origin', 'MarinePelagicBiomes', 'OceanAndSeaRegions', 'MarinePelagicProvince'], axis=1)
Sample_group= Sample_group.transpose()
Sample_group.columns = Sample_group.iloc[0]
Sample_group.drop(Sample_group.index[[0]], inplace=True)
Sample_group.reset_index(inplace=True)
Sample_group = Sample_group.rename(columns = {'index':'Genus'})
Sample_group.head()
```

### Sample to Region mapping

```python
Sample_mapping= Genus_Meta.drop(['Year', 'Month'], axis=1)
Sample_mapping= Sample_mapping.groupby(['Layer of Origin', 'MarinePelagicBiomes', 'OceanAndSeaRegions', 'MarinePelagicProvince', 'SampleID']).sum()
Sample_mapping= Sample_mapping.transpose()
Sample_mapping.reset_index(inplace=True)
Sample_mapping = Sample_mapping.rename(columns = {'index':'Genus'})
Sample_mapping.head()
```

```python
Sample_melt = Sample_mapping.melt(id_vars='Genus')
Sample_melt.head()
```

```python
fig10 = px.treemap(Sample_melt, path=['Layer of Origin', 'MarinePelagicBiomes', 'OceanAndSeaRegions', 'MarinePelagicProvince', 'SampleID'], values='value', color= 'Layer of Origin')
fig10.show()
```

## Coordinated Multiple views

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1(children='TARA Ocean Microbiome Analysis'),
    dcc.Graph(id='Overview-graph', figure=fig2),
    dcc.Dropdown(id='Domain-selector',
                 options= taxonomy['Domain'].unique()
                ),
    dcc.Graph(id='Phylum-graph'),
    dcc.Dropdown(id='Phylum-selector',
                 options= taxonomy['Phylum'].unique()
                ),
    dcc.Graph(id='Sample-graph')
    
])

@app.callback(
    dash.dependencies.Output('Phylum-graph', 'figure'), 
    [dash.dependencies.Input('Domain-selector', 'value')])

def update_graph1(Domain):
    dff= taxonomy[taxonomy['Domain']== Domain]
    Domain_Phylum= dff.groupby(['Domain', 'Phylum']).agg(['sum'])
    Domain_Phylum['Percent']= Domain_Phylum.sum(axis=1, numeric_only=True) / 139
    Domain_Phylum.sort_values(by=['Percent'], inplace=True, ascending=False)
    Domain_Phylum.reset_index(inplace=True)
    Fig1 = px.bar(Domain_Phylum, x="Domain", y= 'Percent', color= 'Phylum', text_auto=True)
    return Fig1       


@app.callback(
    dash.dependencies.Output('Sample-graph', 'figure'), 
    [dash.dependencies.Input('Phylum-selector', 'value')])

def update_graph2(Phylum):
    dff= taxonomy[taxonomy['Phylum']== Phylum]
    Sample= dff.drop(['Domain', 'Class', 'Order', 'Family', 'Genus', 'OTU.rep'], axis=1)
    Sample= Sample.groupby(['Phylum']).sum()
    Sample= Sample.transpose()
    Sample.reset_index(inplace=True)
    Sample = Sample.rename(columns = {'index':'SampleID'})
    Fig2 = px.line(Sample, x= 'SampleID', y= Sample.columns )
    return Fig2 
```

```python
app.run_server()
```
